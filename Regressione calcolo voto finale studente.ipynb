## Import Modules

```
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

## Load Data

```
data = pd.read_csv("student_marks.csv", sep=",")
print(data.head())
```

## Separate Data

```
predict = "final"

x = np.array(data.drop([predict], axis = 1)) # aggiunto axis = 1
y = np.array(data[predict])

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.1)
```

## Define Model

```
linear = linear_model.LinearRegression()

linear.fit(x_train, y_train)
acc = linear.score(x_test, y_test)

print(acc)
```

## Constants of "Best-Fit" Equation

```
print('Coefficient: \n', linear.coef_) # y = mx + c da x1 x2 x3 x4 x5
print('Intercept: \n', linear.intercept_)
```

## Predictions for Specific Students


```
predictions = linear.predict(x_test)

for x in range(len(predictions)):
  print(f'Predicted value:{predictions[x]:0.1f}\t inputs: {x_test[x]}\t Actual Output:{y_test[x]}')
```
